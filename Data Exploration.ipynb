### Diverging Bar Chart:remote work frequency & satisfaction score

```python
remote_satisfaction = df.groupby('Remote_Work_Frequency')['Employee_Satisfaction_Score'].mean()

remote_satisfaction_diff = remote_satisfaction - remote_satisfaction.mean()

plt.figure(figsize=(10, 6))
colors = ['red' if x < 0 else 'green' for x in remote_satisfaction_diff]
remote_satisfaction_diff.plot(kind='bar', color=colors)
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Impact of Remote Work Frequency on Satisfaction')
plt.xlabel('Remote Work Frequency')
plt.ylabel('Difference from Average Satisfaction')
plt.show()
```

Highest Satisfaction (25% Remote Work Frequency): The bar at 25% remote work frequency shows the highest positive deviation, meaning employees working remotely 25% of the time have significantly higher satisfaction compared to the average.

Lowest Satisfaction (0% and 100% Remote Work Frequency): Both 0% and 100% remote work frequencies show negative deviations, suggesting that employees with no remote work or full-time remote work tend to have lower satisfaction than the average.

By observing the resignation distribution of employees with different education levels, we can analyze the relationship between education level and employee resignation rate, which is helpful to understand the relationship between performance indicators and retention/resignation as well as the influence of demographic factors.

```python
Department_resignation = df.groupby(['Department', 'Resigned']).size().unstack()
percentage = Department_resignation.div(Department_resignation.sum(axis=1), axis=0) * 100

ax = Department_resignation.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])

for i, (index, row) in enumerate(Department_resignation.iterrows()):
    cumulative = 0
    for j, value in enumerate(row):
        percent = percentage.iloc[i, j]
        cumulative += value
        plt.text(
            x=i,
            y=cumulative - value / 2,
            s=f'{percent:.1f}%',
            ha='center',
            va='center',
            fontsize=9,
            color='black'
        )
plt.title('Resignation by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Resigned', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.show()
```

```python
Job_resignation = df.groupby(['Job_Title', 'Resigned']).size().unstack()
percentage = Job_resignation.div(Job_resignation.sum(axis=1), axis=0) * 100

ax = Job_resignation.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])

for i, (index, row) in enumerate(Job_resignation.iterrows()):
    cumulative = 0
    for j, value in enumerate(row):
        percent = percentage.iloc[i, j]
        cumulative += value
        plt.text(
            x=i,
            y=cumulative - value / 2,
            s=f'{percent:.1f}%',
            ha='center',
            va='center',
            fontsize=9,
            color='black'
        )
plt.title('Resignation by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.legend(title='Resigned', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.show()
```
